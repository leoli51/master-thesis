# Explore Random Sample data

1. Check distribution over time to check whether the sample is more or less uniform. 

```python
import json

import seaborn as sns

from project.youtube.models import SearchResultItem
```

```python
with open("data/random_sample_1000.json", "r") as sample_file:
	samples_raw = json.load(sample_file)
	samples = [SearchResultItem.from_dict(sample_raw) for sample_raw in samples_raw]

sns.histplot([sample.published_at.date() for sample in samples])
print(len(set([sample.id for sample in samples])))
print(len(set([sample.channel_id for sample in samples])))
```
